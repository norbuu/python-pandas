```python
#loading Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import the data

data = pd.read_csv('./fifa_data.csv')

df = pd.DataFrame(data)

df.head(5)
```

```python
plt.figure(figsize = (8,5))

plt.hist(df['Overall'], bins = 100, alpha = 0.5)

plt.xlabel('Skill level')

plt.ylabel('Number of players')

plt.title('Distribution of player skills (Overall)', fontsize = 12)

plt.show()
```

```python
left = df['Preferred Foot'][df['Preferred Foot'] == 'Left'].count()

right = df['Preferred Foot'][df['Preferred Foot'] == 'Right'].count()
```

```python
labels = ['Left', 'Right']

plt.figure(figsize = (8,5))

plt.pie([left,right], labels = labels, autopct = '%.2f %%')

plt.title('Foot Preference of FIFA Players', fontsize = 12)

plt.show()
```

```python
df['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in df['Weight']]

df['Weight']

light = df.loc[df['Weight'] < 125].count()[0]

light_medium = df.loc[(df['Weight'] >= 125) &(df['Weight'] < 150)].count()[0]

medium = df.loc[(df['Weight'] >=150) & (df['Weight'] <175)].count()[0]

medium_heavy = df.loc[(df['Weight'] >= 175) & (df['Weight'] < 200)].count()[0]

heavy = df.loc[df['Weight'] >= 200].count()[0]
```

```python
plt.style.use('ggplot')

plt.figure(figsize = (8,5))

plt.title('Weight Distribution of Fifa Players (in lbs)', fontsize = 12)

weights = [light, light_medium, medium, medium_heavy, heavy]

labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (.3,.2,0,0,.3)

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode = explode)

plt.show()
```

```python
barcelona = df.loc[df['Club'] == 'FC Barcelona']['Overall']

madrid = df.loc[df['Club'] == 'Real Madrid']['Overall']

legia = df.loc[df['Club'] == 'Legia Warszawa']['Overall']
```

```python
plt.figure(figsize = (5,8))

labels = ['FC Barcelona', 'Real Madrid', 'Legia Warszawa']

boxes = plt.boxplot([barcelona, madrid, legia], labels = labels, patch_artist = True, medianprops = {'linewidth' : 2})

plt.title('Team Comparision', fontsize = 12)

plt.ylabel('FIFA Overall Rating')

for box in boxes['boxes']:
    #Set edge color
    box.set(color = '#4286f4', linewidth = 2)

    #Change Fill Color
    box.set(facecolor = '#e0e0e0')
    
plt.show()
```
