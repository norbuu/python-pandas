```python
from IPython.display import display

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('pokemon_data.csv')
df = pd.DataFrame(data)
df.head(5)
```

```python
df.columns
```

```python
df.columns = [x.replace(".","").replace(" ", "_") for x in df.columns]
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
```

```python
gen_1 = df.loc[(df['Generation'] == 1) & (~df['Name'].str.contains('Mega'))]
gen_1 = gen_1.sort_values('Total', ascending=False).reset_index(drop=True)
gen_1.head(5)
```

```python
plot_1 = gen_1.iloc[0:5].plot(x="Name", y = "Total", color = 'blue', kind="bar", alpha= 0.5, title = "Total stats")
```

```python
top_3 = df.iloc[:, 2:10].groupby(['Type_1']).mean().sort_values('Attack', ascending= False).round(2).head(3)
top_3
```

```python
plot_2 = top_3.plot(kind = 'bar', title = 'Top 3 by attack', alpha = 0.5).legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
gen_1['Count'] = 1
gen_1.groupby('Type_1').count().sort_values('Count', ascending=False)['Count']
```
